# Official MovieLens Demo

This notebook runs the canonical Phase 1 demo for the recommender behavior QA tool.

It compares:
- `Model A`: popularity baseline
- `Model B`: genre-profile recommender with popularity prior

The goal is to show that aggregate offline ranking metrics are useful, but they can hide bucket-level and behavioral tradeoffs.

```python
from pathlib import Path
import sys

cwd = Path.cwd().resolve()
candidate_roots = [
    cwd,
    cwd.parent,
    cwd / 'studies' / '01-recommender-offline-eval',
]
project_root = next(
    path
    for path in candidate_roots
    if (path / 'src' / 'recommender_offline_eval').exists()
)
sys.path.insert(0, str(project_root / 'src'))

from recommender_offline_eval import load_evaluation_config, run_canonical_demo, run_evaluation
```

```python
result = run_canonical_demo(
    data_dir=project_root / 'data',
    output_dir=project_root / 'output',
)
aggregate_table = result['aggregate_table']
bucket_table = result['bucket_table']
behavior_table = result['behavior_table']
result['report_path'], result['metrics_path'], result['chart_path']
```
